```python
import json

# Load data from annotated_posts.json
with open('comments/annotated_posts.json', 'r', encoding='utf-8') as file:
    annotated_posts = json.load(file)

# Display the loaded data
print(annotated_posts)
```

```python
# Filter out posts where message is null
filtered_posts = [post for post in annotated_posts if post.get('message') is not None]

# Display the filtered posts
#print(filtered_posts)

# Filter posts where category is politics
filtered_politics_posts = [post for post in filtered_posts if post.get('category') == 'politics']

# Display the filtered politics posts
print(filtered_politics_posts)

# for each post, append a new field with SHA256 hash of the url. Prefix the hash with 'comment_'
import hashlib
for post in filtered_politics_posts:
    if 'url' in post and post['url'] is not None:
        url = post['url']
        # Create SHA256 hash of the URL
        sha256_hash = hashlib.sha256(url.encode()).hexdigest()
        # Prefix the hash with 'comment_'
        post['comment_id'] = f'comments_{sha256_hash}.json'
    else:
        post['comment_id'] = None


# Save the filtered posts to a new JSON file
with open('comments/filtered_politics_posts.json', 'w', encoding='utf-8') as file:
    json.dump(filtered_politics_posts, file, ensure_ascii=False, indent=4)
```

```python
# create a new dir for comment files belonging to the posts
import os
import shutil
import json
import hashlib
import re

# Create a new directory for comment files if it doesn't exist
if not os.path.exists('politics_comments'):
    os.makedirs('politics_comments')

# Move the comment files to the new directory
for post in filtered_politics_posts:
    if 'comment_id' in post and post['comment_id'] is not None:
        # Extract the comment file name from the comment_id
        comment_file_name = post['comment_id']
        # Create the full path for the comment file
        comment_file_path = os.path.join('comments', comment_file_name)
        # Check if the comment file exists
        if os.path.exists(comment_file_path):
            # Move the comment file to the new directory
            shutil.move(comment_file_path, os.path.join('politics_comments', comment_file_name))
        else:
            print(f"Comment file {comment_file_path} does not exist. Skipping.")
```
